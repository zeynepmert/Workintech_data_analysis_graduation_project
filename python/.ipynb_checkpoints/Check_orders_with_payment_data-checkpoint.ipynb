orders_update.csv dosyasindaki duzeltmeleri yapip olusturdugum orders_with_payment.csv dosyamizin dogrulugunu kontrol ettim.

```python
import pandas as pd
```

```python
orders_df = pd.read_csv('../data/orders_updated.csv')
sales_df = pd.read_csv('../data/sales.csv')
orders_with_payment_df = pd.read_csv('../data/orders_with_payment.csv')
```

```python
# customer_id = 1 olan musteriyi sectim cunku bu musteri birden fazla alisveris yapmis.
customer_1_orders = orders_df[orders_df['customer_id'] == 1]
display(customer_1_orders['order_id'])
```

```python
# customer = 1 'e ait order_id'ler : 455, 670, 729 --> Bunu eski data(orders_update)'dan kontrol ettik.
```

```python
# customer_id'si 1 olan musterinin her yaptigi order_id icin total price'i
customer_1_orders = orders_with_payment_df [orders_with_payment_df ['customer_id'] == 1]
display(customer_1_orders[['order_id', 'payment']])
```

```python
# sales tablosunu kullanarak order_id'si 455, 670 ve 729 olan orderlarin total_price degerleri
selected_order_ids = [455, 670, 729]
selected_orders = sales_df[sales_df['order_id'].isin(selected_order_ids)]
display(selected_orders[['order_id', 'total_price']])
```

```python
# orders_with_payment.csv'den customer_id'si 1 olan musterinin payment degeri
customer_1_payment = orders_with_payment_df[orders_with_payment_df['customer_id'] == 1]
display(customer_1_payment[['customer_id', 'payment']])
```
