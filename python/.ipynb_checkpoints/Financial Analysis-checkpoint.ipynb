```python
import pandas as pd
import numpy as np
sales = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\sales.csv")
orders = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\orders.csv")
customers = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\customers.csv")
products = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\products.csv")
sales
```

```python
total_revenue = sales["total_price"].sum()
print(f"Toplam Ciro: ${total_revenue:,.2f}")
```

```python
sales_orders = pd.merge(sales, orders, on="order_id", how="left")
```

```python
sales_orders_customers = pd.merge(sales_orders, customers, on="customer_id", how="left")
```

```python
full_data = pd.merge(sales_orders_customers, products, left_on="product_id", right_on="product_ID", how="left")
```

```python
full_data["order_date"] = pd.to_datetime(full_data["order_date"])
#monthly revenue
month_revenue=full_data.groupby(full_data["order_date"].dt.to_period("M"))["total_price"].sum()
year_revenue=full_data.groupby(full_data["order_date"].dt.year)["total_price"].sum()
print(month_revenue)
print(year_revenue)
```

```python
print(orders)
```

```python
product_revenue=full_data.groupby("product_name")["total_price"].sum().sort_values(ascending=False)
print(product_revenue)
```

```python
customer_revenue = full_data.groupby("customer_name")["total_price"].sum().sort_values(ascending=False)
print(customer_revenue)
```

```python
basket_amount= full_data.groupby("order_id")["total_price"].sum()
avg_basket_amount= basket_amount.mean()
print(f"Avg. Basket: ${avg_basket_amount:.2f}")
```

```python
full_data["profit"] = full_data["total_price"] - (full_data["quantity_x"] * full_data["price"])
total_profit = full_data["profit"].sum()
print(f"Total profit: ₺{total_profit:,.2f}")
```
