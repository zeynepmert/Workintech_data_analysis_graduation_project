```python
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, hspace=0.35, wspace=0.3)

# 1. Kategori x Renk Heatmap
ax1 = fig.add_subplot(gs[0, :])
category_color = df.groupby(['product_type', 'colour'])['quantity_sold'].sum().unstack(fill_value=0)
top_15_colors = df.groupby('colour')['quantity_sold'].sum().nlargest(15).index
category_color_filtered = category_color[category_color.columns.intersection(top_15_colors)]
sns.heatmap(category_color_filtered, annot=True, fmt='g', cmap='YlOrRd',linewidths=0.5, ax=ax1, cbar_kws={'label': 'Sat Miktar'})
ax1.set_title('Kategori x Renk Sat Matrisi (Top 15 Renk)',fontsize=12, fontweight='bold')
ax1.set_xlabel('Renk')
ax1.set_ylabel('Kategori')

# 2. Top 15 En Popler Renk
ax2 = fig.add_subplot(gs[1, 0])
top_colors = df.groupby('colour')['quantity_sold'].sum().nlargest(10)
top_colors.plot(kind='bar', ax=ax2, color='coral')
ax2.set_title('En Popler 10 Renk')
ax2.tick_params(axis='x', rotation=45)

plt.suptitle('RENK TERCHLER ANALZ', fontsize=18, fontweight='bold', y=0.99, x=0.43)
plt.show()
```

```python
import seaborn as sns
import numpy as np

fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, hspace=0.35, wspace=0.3)

# 1. Kategori x Beden Heatmap
ax1 = fig.add_subplot(gs[0, :])
category_size = df.groupby(['product_type', 'size'])['quantity_sold'].sum().unstack(fill_value=0)
sns.heatmap(category_size, annot=True, fmt='g', cmap='Blues', 
            linewidths=0.5, ax=ax1, cbar_kws={'label': 'Sat Miktar'})
ax1.set_title('Kategori x Beden Sat Matrisi', 
              fontsize=12, fontweight='bold')
ax1.set_xlabel('Beden')
ax1.set_ylabel('Kategori')

# 2. Top 15 En Popler Beden
ax2 = fig.add_subplot(gs[1, 0])
top_sizes = df.groupby('size')['quantity_sold'].sum().nlargest(15).sort_values()
top_sizes.plot(kind='barh', ax=ax2, color='lightcoral')
ax2.set_title('Top 15 En Popler Beden', fontsize=12, fontweight='bold')
ax2.set_xlabel('Sat Miktar')

# 3. Kategorilerde Beden Dalm (Stacked Bar)
ax3 = fig.add_subplot(gs[1, 1])
top_10_sizes = df.groupby('size')['quantity_sold'].sum().nlargest(10).index
category_size_pivot = df.groupby(['product_type', 'size'])['quantity_sold'].sum().unstack(fill_value=0)
category_size_pivot[top_10_sizes].plot(kind='bar', stacked=True, ax=ax3, 
                                       colormap='Paired')
ax3.set_title('Kategorilerde Beden Dalm (Top 10 Beden)', fontsize=12, fontweight='bold')
ax3.set_xlabel('Kategori')
ax3.set_ylabel('Toplam Sat')
ax3.tick_params(axis='x', rotation=45)
ax3.legend(title='Beden', fontsize=8, title_fontsize=9, bbox_to_anchor=(1.05, 1))

plt.suptitle('BEDEN TERCHLER ANALZ',fontsize=18, fontweight='bold', y=0.995)
plt.show()
```
