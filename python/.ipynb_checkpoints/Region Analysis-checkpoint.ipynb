```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Verileri tanımlama
sales = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\sales.csv")
orders = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\orders.csv")
customers = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\customers.csv")
products = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\products.csv")
```

```python
# Verileri date_time çevirme
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['delivery_date'] = pd.to_datetime(orders['delivery_date'])
```

```python
#sales ve orders tabloları joinleme
df = pd.merge(sales, orders, on='order_id', how='left')
```

```python
# customers tablosunuda ekleme
df = pd.merge(df, customers, on='customer_id', how='left')
```

```python
df = df.sort_values('order_date')

df.head()
```

```python
# Şehre göre gelir hesaplama ve grafik
revenue_city = df.groupby('state')['total_price'].sum().reset_index().sort_values(by='total_price', ascending=False)

# İlk 10 şehir görselleştirme
plt.figure(figsize=(12,5))
sns.barplot(data=revenue_city.head(10), x='state', y='total_price')
plt.title('Top 10 Cities by Revenue')
plt.ylabel('Total Revenue')
plt.xlabel('State')
plt.xticks(rotation=30)
plt.show()
```

```python
summary = df.groupby(['state', 'city']).agg(
    Total_Revenue=('total_price', 'sum'),
    Order_Count=('order_id', 'nunique')
).reset_index()

summary
```

```python
#Orders by Region / State
orders_city = df.groupby('state')['order_id'].nunique().reset_index().sort_values(by='order_id', ascending=False)
plt.figure(figsize=(12,5))
sns.barplot(data=orders_city.head(10), x='state', y='order_id')
plt.title('Number of Orders by State')
plt.ylabel('Order Count')
plt.xlabel('State')
plt.xticks(rotation=30)
plt.show()
```

```python
#Average Spend by Region
avg_spend_region = df.groupby('state').agg(
    total_revenue=('total_price', 'sum'),
    customer_count=('customer_id', 'nunique')
).reset_index()

avg_spend_region['avg_spend'] = avg_spend_region['total_revenue'] / avg_spend_region['customer_count']

# Görselleştirme
plt.figure(figsize=(12,5))
sns.barplot(data=avg_spend_region.sort_values('avg_spend', ascending=False).head(10),
            x='state', y='avg_spend')
plt.title('Average Spend per Customer by State')
plt.xticks(rotation=45)
plt.show()
```

```python
# Delivery Time by Region
df['delivery_duration'] = (df['delivery_date'] - df['order_date']).dt.days

delivery_time_region = df.groupby('state')['delivery_duration'].mean().reset_index().sort_values('delivery_duration')

plt.figure(figsize=(12,5))
sns.barplot(data=delivery_time_region, x='state', y='delivery_duration')
plt.title('Average Delivery Time by Region')
plt.ylabel('Average Delivery (days)')
plt.xticks(rotation=45)
plt.show()
```

```python
# KPI özetleri
summary_kpis = pd.DataFrame({
    'Total Revenue ($)': [df['total_price'].sum()],
    'Total Orders': [df['order_id'].nunique()],
    'Unique Customers': [df['customer_id'].nunique()],
    'Average Order Value ($)': [df['total_price'].sum() / df['order_id'].nunique()],
    'Avg Delivery Time (days)': [df['delivery_duration'].mean()]
})
summary_kpis
```
