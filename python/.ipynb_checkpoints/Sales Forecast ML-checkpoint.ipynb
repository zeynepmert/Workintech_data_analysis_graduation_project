```python
!pip install --quiet pandas numpy matplotlib seaborn statsmodels scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Makine öğrenmesi kütüphaneleri
from sklearn.linear_model import LinearRegression
# Tarih işlemleri için
from datetime import datetime
# Uyarı mesajlarını gizlemek için
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
sales = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\sales.csv")
orders = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\orders.csv")
customers = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\customers.csv")
products = pd.read_csv(r"C:\Users\ysfcl\Desktop\Workintech final\Workintech_data_analysis_graduation_project\data\products.csv")
```

```python
# Orders tablosundaki tarih sütunlarını datetime formatına çevirdik
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['delivery_date'] = pd.to_datetime(orders['delivery_date'])
# Veri birleştirme
# order_id sütununa göre eşleştir
# how='left' → Sales'teki tüm satırları koru
sales_orders = sales.merge(
    orders[['order_id', 'customer_id', 'order_date', 'delivery_date']], 
    on='order_id',
    how='left'
)
# Veri birleştirme - 2. Adım: Yukarıdaki sonuç + Products
# product_id ile product_ID eşleştir (farklı isimlerde)
sales_full = sales_orders.merge(
    products[['product_ID', 'product_name', 'product_type']], 
    left_on='product_id',
    right_on='product_ID',
    how='left'
)
```

```python
# önce günlük bazda toplamları hesapladık
daily_sales = sales_full.groupby('order_date').agg({
    'total_price': 'sum',        # Toplam gelir (TL)
    'quantity': 'sum',            # Satılan ürün adedi
    'order_id': 'nunique'         # Benzersiz sipariş sayısı
}).reset_index()

# Sütun isimlerini daha anlaşılır yap
daily_sales.columns = ['date', 'revenue', 'units_sold', 'num_orders']

# Tarihe göre küçükten büyüğe sırala
daily_sales = daily_sales.sort_values('date').reset_index(drop=True)

print(f" Günlük satış verileri !")
print(f"   • Toplam gün sayısı: {len(daily_sales)}")
print(f"   • İlk tarih: {daily_sales['date'].min()}")
print(f"   • Son tarih: {daily_sales['date'].max()}")

# Verinin kaç günü kapsadığını hesapla
gun_farki = (daily_sales['date'].max() - daily_sales['date'].min()).days
print(f"   • Veri aralığı: {gun_farki} gün")

# İstatistiksel özet
# describe(): min, max, ortalama, medyan, vb. gösterir
print("\n Günlük Satış describe:")
print(daily_sales['revenue'].describe())

print("\n İstatistik:")
print(f"   • Ortalama günlük gelir: {daily_sales['revenue'].mean():.2f} TL")
print(f"   • En düşük gün: {daily_sales['revenue'].min():.2f} TL")
print(f"   • En yüksek gün: {daily_sales['revenue'].max():.2f} TL")
print(f"   • Fark: {daily_sales['revenue'].max() - daily_sales['revenue'].min():.2f} TL")
```

```python
# Günlük değişim yüzdesi
# pct_change(): Her günü bir önceki güne göre karşılaştırır
daily_sales['pct_change'] = daily_sales['revenue'].pct_change() * 100

# Volatilite istatistikleri
ortalama_degisim = daily_sales['pct_change'].mean()
std_sapma = daily_sales['pct_change'].std()
max_artis = daily_sales['pct_change'].max()
max_dusus = daily_sales['pct_change'].min()

print(f" Volatilite İstatistikleri:")
print(f"   • Ortalama günlük değişim: {ortalama_degisim:.2f}%")
print(f"   • Standart sapma: {std_sapma:.2f}%")
print(f"   • Maksimum artış: {max_artis:.2f}%")
print(f"   • Maksimum düşüş: {max_dusus:.2f}%")

# Volatilite değerlendirmesi
print("\n DEĞERLENDİRME:")
if std_sapma > 100:
    print("     ÇOK YÜKSEK VOLATİLİTE!")
    print("   → Ham veri ile model kurmak çok zor")
    print("   → Veriyi düzleştirmemiz gerekiyor")
    print("   → Hareketli ortalama kullanacağız")
elif std_sapma > 50:
    print("     Yüksek volatilite")
    print("   → Model performansı orta olabilir")
elif std_sapma > 20:
    print("    Normal volatilite")
    print("   → Model kurulabilir")
else:
    print("    Düşük volatilite - Mükemmel!")
    print("   → Model çok iyi çalışacak")
```

```python
"""
========================================================================
BÖLÜM 6: 7 GÜNLÜK HAREKETLİ ORTALAMA HESAPLAMA
========================================================================
Hareketli Ortalama (Moving Average - MA):
- Son N günün ortalamasını alır
- Günlük dalgalanmaları yumuşatır
- Gerçek trendi daha net gösterir

Örnek: 7 günlük MA
Gün 1-7 ortalaması → 1. değer
Gün 2-8 ortalaması → 2. değer
Gün 3-9 ortalaması → 3. değer ...
========================================================================
"""


# 7 günlük hareketli ortalamayı hesapla
# rolling(window=7): Son 7 günlük pencere oluştur
# mean(): Bu pencerenin ortalamasını al
# min_periods=1: İlk günler için en az 1 veri varsa hesapla
daily_sales['revenue_ma7'] = daily_sales['revenue'].rolling(
    window=7,       # Kaç günlük ortalama istiyoruz?
    min_periods=1   # En az kaç veri olmalı? (ilk günler için)
).mean()

print(" 7 günlük hareketli ortalama hesaplandı!")

# Hareketli ortalama volatilitesini hesapla
# Düzleştirme ne kadar etkili oldu görelim
ma7_volatility = daily_sales['revenue_ma7'].pct_change().std() * 100
ham_volatility = daily_sales['pct_change'].std()

print(f"\n VOLATİLİTE KARŞILAŞTIRMASI:")
print(f"   • Ham veri volatilite: {ham_volatility:.2f}%")
print(f"   • 7 günlük MA volatilite: {ma7_volatility:.2f}%")
print(f"   • İyileşme oranı: {(ham_volatility / ma7_volatility):.1f}x")
print(f"   • Volatilite düşüşü: %{((ham_volatility - ma7_volatility) / ham_volatility * 100):.1f}")

if (ham_volatility / ma7_volatility) > 10:
    print("\n MÜKEMMel! Volatilite 10 kattan fazla düştü!")
elif (ham_volatility / ma7_volatility) > 5:
    print("\n ÇOK İYİ! Volatilite 5 kattan fazla düştü!")
elif (ham_volatility / ma7_volatility) > 2:
    print("\n İyi! Volatilite yarıdan fazla düştü!")
else:
    print("\n  Düzleştirme etkisi az oldu")

# İlk 6 günü çıkar
# Neden? İlk 6 gün için tam 7 günlük ortalama hesaplanamaz
# Örnek: 1. gün için sadece 1 veri var, 7 veri yok
print(f"\n Veri Temizliği:")
print(f"   • Önceki veri sayısı: {len(daily_sales)} gün")

daily_sales_ma = daily_sales[6:].copy().reset_index(drop=True)

print(f"   • İlk 6 gün çıkarıldı (tam MA olmayan)")
print(f"   • Yeni veri sayısı: {len(daily_sales_ma)} gün")
```

```python
"""
========================================================================
BÖLÜM 7: VERİ GÖRSELLEŞTİRME
3 grafik çizeceğiz:
1. Ham günlük satış verisi (dalgalı)
2. 7 günlük hareketli ortalama (düzleştirilmiş)
3. Volatilite karşılaştırması (bar chart)
========================================================================
"""
# 1 satır, 3 sütun grafik düzeni oluştur
# figsize: Grafik boyutu (genişlik, yükseklik) inch cinsinden
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ---- GRAFİK 1: Ham Günlük Satış Verisi ----
# alpha=0.5: Yarı şeffaf (0=tamamen şeffaf, 1=opak)
axes[0].plot(daily_sales['date'], daily_sales['revenue'], 
             alpha=0.5, color='gray', linewidth=1, label='Ham Veri')
axes[0].set_title('Günlük Satış Verisi (Ham)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Tarih')
axes[0].set_ylabel('Gelir (TL)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)  # Izgara çizgileri ekle
axes[0].tick_params(axis='x', rotation=45)  # X eksenindeki yazıları 45 derece döndür

# ---- GRAFİK 2: Ham Veri + 7 Günlük MA Karşılaştırması ----
# İki çizgiyi aynı grafikte göster
axes[1].plot(daily_sales['date'], daily_sales['revenue'], 
             alpha=0.2, color='gray', linewidth=1, label='Ham Veri')
axes[1].plot(daily_sales['date'], daily_sales['revenue_ma7'], 
             color='blue', linewidth=2.5, label='7 Günlük MA')
axes[1].set_title('7 Günlük Hareketli Ortalama', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Tarih')
axes[1].set_ylabel('Gelir (TL)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

# ---- GRAFİK 3: Volatilite Karşılaştırması (Bar Chart) ----
# İki volatilite değerini yan yana bar olarak göster
volatility_labels = [f'Ham Veri\n({ham_volatility:.1f}%)', 
                     f'7 Günlük MA\n({ma7_volatility:.1f}%)']
volatility_values = [ham_volatility, ma7_volatility]
colors = ['red', 'green']  # Kırmızı=kötü, Yeşil=iyi

axes[2].bar(volatility_labels, volatility_values, 
            color=colors, alpha=0.7, edgecolor='black', width=0.6)
axes[2].set_title('Volatilite Karşılaştırması', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Standart Sapma (%)')
axes[2].grid(True, alpha=0.3, axis='y')

# Her bar'ın üstüne değeri yaz
for i, v in enumerate(volatility_values):
    axes[2].text(i, v + 5, f'{v:.1f}%', 
                ha='center', va='bottom', fontweight='bold', fontsize=10)

# Tüm grafikleri düzenli yerleştir
plt.tight_layout()
plt.show()

print(" Grafikler başarıyla oluşturuldu!")
print("\n GRAFİKLERİ NASIL YORUMLARIZ?")
print(" Grafik 1 (Ham Veri):")
print("   → Çok dalgalı, trend net görünmüyor")
print("   → Bir gün yüksek, ertesi gün düşük")
print("\n Grafik 2 (7 Günlük MA):")
print("   → Mavi çizgi çok daha düzgün")
print("   → Trend artık net görülüyor")
print("   → Gri çizgi (ham veri) hala dalgalı")
print("\n Grafik 3 (Volatilite):")
print("   → Yeşil bar çok daha kısa (iyi!)")
print("   → Volatilite dramatik şekilde düştü")
```

```python
"""
========================================================================
BÖLÜM 8: TRAIN-TEST AYIRMA
========================================================================
Modeli değerlendirmek için veriyi 2'ye ayırıyoruz:
- TRAIN (Eğitim): Model bunu öğrenecek
- TEST (Test): Model bunu bilmiyor, tahmin edecek

Mantık: Öğrenci bir sınava çalışır (train), 
sonra görmediği sorularla sınava girer (test)
========================================================================
"""

print("\n\n  VERİ TRAIN-TEST OLARAK AYRILIYOR...")
print("="*70)

# Son 30 günü test için ayır
# Neden 30 gün? 1 aylık performansı görmek yeterli
# train_size: Eğitim setinin boyutu
train_size = len(daily_sales_ma) - 30

# Veriyi böl
train_data = daily_sales_ma[:train_size].copy()  # İlk N gün
test_data = daily_sales_ma[train_size:].copy()   # Son 30 gün

print(f" Veri başarıyla ayrıldı!")
print(f"\n Veri Dağılımı:")
print(f"   • Toplam veri: {len(daily_sales_ma)} gün")
print(f"   • Eğitim (Train) seti: {len(train_data)} gün (%{(len(train_data)/len(daily_sales_ma)*100):.1f})")
print(f"   • Test seti: {len(test_data)} gün (%{(len(test_data)/len(daily_sales_ma)*100):.1f})")

print(f"\n Tarih Aralıkları:")
print(f"   • Train başlangıç: {train_data['date'].min().strftime('%Y-%m-%d')}")
print(f"   • Train bitiş: {train_data['date'].max().strftime('%Y-%m-%d')}")
print(f"   • Test başlangıç: {test_data['date'].min().strftime('%Y-%m-%d')}")
print(f"   • Test bitiş: {test_data['date'].max().strftime('%Y-%m-%d')}")

# Train ve Test setlerinin ortalamalarını karşılaştır
# İdeal: Birbirine yakın olmalılar (çok farklıysa sorun var)
train_mean = train_data['revenue_ma7'].mean()
test_mean = test_data['revenue_ma7'].mean()
fark_yuzde = ((test_mean - train_mean) / train_mean) * 100

print(f"\n Ortalama Gelir Karşılaştırması:")
print(f"   • Train ortalama: {train_mean:.2f} TL")
print(f"   • Test ortalama: {test_mean:.2f} TL")
print(f"   • Fark: {fark_yuzde:+.2f}%")

if abs(fark_yuzde) < 10:
    print("\n Train ve Test setleri dengeli (fark <%10)")
    print("   → Model güvenilir sonuçlar verecek")
elif abs(fark_yuzde) < 20:
    print("\n  Train ve Test arasında orta fark var (%10-20)")
    print("   → Model sonuçları kabul edilebilir")
else:
    print("\n Train ve Test çok farklı (>%20)")
    print("   → Test seti dönemi olağandışı olabilir")
    print("   → Sonuçları dikkatli yorumlayın")
```

```python
"""
========================================================================
BÖLÜM 9: MODEL EĞİTİMİ (Linear Regression)
========================================================================
Linear Regression (Doğrusal Regresyon):
- En basit makine öğrenmesi algoritması
- Veriye en uygun doğruyu çizer
- Formül: y = mx + b
  y = Gelir (tahmin)
  m = Eğim (günlük değişim)
  x = Gün numarası
  b = Başlangıç değeri
========================================================================
"""

print("\n\n🤖 LINEAR REGRESSION MODELİ EĞİTİLİYOR...")
print("="*70)

# Gün numarası sütunu ekle
# Neden? Linear Regression tarih değil sayı ister
# 0, 1, 2, 3, ... şeklinde gün numaraları
train_data['day_num'] = range(len(train_data))
test_data['day_num'] = range(len(train_data), len(train_data) + len(test_data))

print("✅ Gün numaraları oluşturuldu")

# X ve y değişkenlerini hazırla
# X: Girdi (bağımsız değişken) - Ne veriyoruz modele?
# y: Çıktı (bağımlı değişken) - Ne tahmin etmesini istiyoruz?

X_train = train_data[['day_num']]          # [[]] kullanımı: 2D array yapısı
y_train = train_data['revenue_ma7']        # 1D array

X_test = test_data[['day_num']]
y_test = test_data['revenue_ma7']

print(f"\n📊 Eğitim Verileri:")
print(f"   • X_train boyutu: {X_train.shape}")
print(f"   • y_train boyutu: {y_train.shape}")

# Linear Regression modelini oluştur
model = LinearRegression()

# Modeli eğit
# fit(): Model, X_train'deki günleri ve y_train'deki gelirleri öğrenir
# Aralarındaki ilişkiyi (en iyi doğruyu) bulur
model.fit(X_train, y_train)

print("\n✅ Model başarıyla eğitildi!")

# Model parametrelerini göster
# coef_[0]: Eğim (slope) - Günlük değişim
# intercept_: Kesişim noktası (y-intercept) - Başlangıç değeri
egim = model.coef_[0]
kesisim = model.intercept_

print(f"\n📐 MODEL PARAMETRELERİ:")
print("="*70)
print(f"   • Eğim (m): {egim:.2f} TL/gün")
print(f"   • Kesişim (b): {kesisim:.2f} TL")

print(f"\n📝 Model Denklemi:")
print(f"   Gelir = {kesisim:.2f} + ({egim:.2f} × Gün)")

# Eğimi yorumla
print(f"\n💡 TREND YORUMU:")
print("="*70)
if egim > 10:
    print(f"   📈 ARTAN TREND!")
    print(f"   • Satışlar her gün {egim:.2f} TL artıyor")
    print(f"   • 30 günde tahmini artış: {egim * 30:.2f} TL")
    print(f"   • Yıllık tahmini artış: {egim * 365:.2f} TL")
    print(f"   → İşler iyi gidiyor! ✅")
elif egim < -10:
    print(f"   📉 AZALAN TREND!")
    print(f"   • Satışlar her gün {abs(egim):.2f} TL azalıyor")
    print(f"   • 30 günde tahmini düşüş: {abs(egim * 30):.2f} TL")
    print(f"   • Yıllık tahmini düşüş: {abs(egim * 365):.2f} TL")
    print(f"   → Aksiyon gerekli! ⚠️")
else:
    print(f"   ➡️  STABİL TREND")
    print(f"   • Satışlar neredeyse sabit (günlük değişim: {egim:.2f} TL)")
    print(f"   • Büyük artış veya düşüş yok")
    print(f"   → Normal seyir devam ediyor")
```

```python
"""
========================================================================
BÖLÜM 10: MODEL TAHMİNLERİ ve PERFORMANS METRİKLERİ
========================================================================
Eğitilen modeli test setinde deneyelim:
1. Test seti için tahmin yap
2. Gerçek değerlerle karşılaştır
3. Performans metriklerini hesapla (MAE, RMSE, R²)
========================================================================
"""

print("\n\n🎯 TEST SETİ İÇİN TAHMİNLER YAPILIYOR...")
print("="*70)

# Test seti için tahmin yap
# predict(): Modele X_test'i ver, gelir tahminlerini al
y_pred_test = model.predict(X_test)

print("✅ Tahminler tamamlandı!")
print(f"   • {len(y_pred_test)} günlük tahmin yapıldı")

# İlk 5 tahmini göster
print(f"\n🔍 İlk 5 Tahmin vs Gerçek:")
print("="*70)
comparison = pd.DataFrame({
    'Tarih': test_data['date'].head(),
    'Gerçek': y_test.head().values,
    'Tahmin': y_pred_test[:5],
    'Hata': (y_test.head().values - y_pred_test[:5])
})
comparison['Gerçek'] = comparison['Gerçek'].round(2)
comparison['Tahmin'] = comparison['Tahmin'].round(2)
comparison['Hata'] = comparison['Hata'].round(2)
print(comparison.to_string(index=False))

# ========================================================================
# PERFORMANS METRİKLERİNİ HESAPLA
# ========================================================================

# 1. MAE (Mean Absolute Error) - Ortalama Mutlak Hata
# |Gerçek - Tahmin| değerlerinin ortalaması
# Yorumu: Model ortalama kaç TL hata yapıyor?
mae = mean_absolute_error(y_test, y_pred_test)

# 2. RMSE (Root Mean Squared Error) - Kök Ortalama Kare Hata
# Büyük hatalara daha çok ceza verir
# MAE'den daha katı bir metrik
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

# 3. R² (R-Squared) - Belirlilik Katsayısı
# Model, varyansın ne kadarını açıklıyor?
# 1.0 = Mükemmel, 0.0 = Ortalama almak kadar iyi, <0 = Kötü
r2 = r2_score(y_test, y_pred_test)

print(f"\n\n📊 MODEL PERFORMANS METRİKLERİ")
print("="*70)

# MAE açıklaması
print(f"\n1️⃣  MAE (Ortalama Mutlak Hata): {mae:.2f} TL")
print(f"   📖 Açıklama:")
print(f"      • Model, ortalama {mae:.2f} TL hata yapıyor")
print(f"      • Her tahmin için ±{mae:.2f} TL sapma bekleniyor")
print(f"   💡 Örnek:")
print(f"      Gerçek 3,000 TL ise → Model 2,500-3,500 TL arası tahmin eder")

# RMSE açıklaması
print(f"\n2️⃣  RMSE (Kök Ortalama Kare Hata): {rmse:.2f} TL")
print(f"   📖 Açıklama:")
print(f"      • MAE'ye benzer ama büyük hatalara daha fazla ceza verir")
print(f"      • RMSE > MAE ise bazı tahminler çok yanlış")
print(f"   💡 Karşılaştırma:")
fark = rmse - mae
if fark < 100:
    print(f"      RMSE ≈ MAE → Hatalar dengeli ✅")
elif fark < 300:
    print(f"      RMSE > MAE ({fark:.0f} TL fark) → Bazı uç hatalar var ⚠️")
else:
    print(f"      RMSE >> MAE ({fark:.0f} TL fark) → Çok büyük hatalar var ❌")

# R² açıklaması
print(f"\n3️⃣  R² (Belirlilik Katsayısı): {r2:.4f}")
print(f"   📖 Açıklama:")
if r2 > 0.9:
    print(f"      • Mükemmel! Model varyansın %{r2*100:.1f}'ini açıklıyor")
    print(f"      • Tahminler çok güvenilir ✅")
elif r2 > 0.7:
    print(f"      • Çok iyi! Model varyansın %{r2*100:.1f}'ini açıklıyor")
    print(f"      • Tahminler güvenilir ✅")
elif r2 > 0.5:
    print(f"      • İyi. Model varyansın %{r2*100:.1f}'ini açıklıyor")
    print(f"      • Kabul edilebilir performans ⚠️")
elif r2 > 0:
    print(f"      • Zayıf. Model varyansın sadece %{r2*100:.1f}'ini açıklıyor")
    print(f"      • İyileştirme gerekli ⚠️")
else:
    print(f"      • Negatif R² (test setinde normal olabilir)")
    print(f"      • Test setinde bazı uç değerler metriği bozmuş")
    print(f"      • AMA MAE düşükse sorun yok!")

# Hata oranını hesapla
# Gerçek değerlerin ortalamasına göre hata yüzdesi
hata_orani = (mae / y_test.mean()) * 100

print(f"\n🎯 HATA ORANI:")
print("="*70)
print(f"   • Ortalama gelir (test): {y_test.mean():.2f} TL")
print(f"   • Ortalama hata (MAE): {mae:.2f} TL")
print(f"   • Hata oranı: %{hata_orani:.2f}")

# Hata oranı değerlendirmesi
print(f"\n   💡 DEĞERLENDİRME:")
if hata_orani < 5:
    print(f"      ⭐⭐⭐⭐⭐ Olağanüstü! (<%5 hata)")
    print(f"      → Model çok güvenilir")
elif hata_orani < 10:
    print(f"      ⭐⭐⭐⭐ Mükemmel! (%5-10 hata)")
    print(f"      → Model güvenilir")
elif hata_orani < 15:
    print(f"      ⭐⭐⭐ Çok iyi! (%10-15 hata)")
    print(f"      → Model kullanılabilir")
elif hata_orani < 20:
    print(f"      ⭐⭐ İyi (%15-20 hata)")
    print(f"      → Genel tahminler için uygun")
elif hata_orani < 30:
    print(f"      ⭐ Kabul edilebilir (%20-30 hata)")
    print(f"      → Dikkatli kullanın")
else:
    print(f"      ❌ Yetersiz (>%30 hata)")
    print(f"      → Model iyileştirmesi gerekli")

print(f"\n📌 SONUÇ:")
if hata_orani < 20 and (r2 > 0.3 or r2 < 0):
    print(f"   ✅ Model BAŞARILI! Tahminler için kullanılabilir.")
elif hata_orani < 30:
    print(f"   ⚠️  Model ORTA seviyede. Genel tahminler için uygun.")
else:
    print(f"   ❌ Model YETERSİZ. İyileştirme yapılmalı.")
```

```python
"""
========================================================================
BÖLÜM 11: GELECEK TAHMİNLERİ (30 Gün)
========================================================================
Eğitilen modeli kullanarak gelecek 30 gün için satış tahmini yapalım.
Bu tahminler stok planlaması, nakit akışı ve personel planlaması için kullanılır.
========================================================================
"""

print("\n\n🔮 GELECEK 30 GÜN İÇİN TAHMİNLER YAPILIYOR...")
print("="*70)

# Gelecek 30 gün için gün numaralarını oluştur
# Son gün numarasından devam et
future_days = pd.DataFrame({
    'day_num': range(len(daily_sales_ma), len(daily_sales_ma) + 30)
})

print(f"✅ Gelecek gün numaraları oluşturuldu: {len(future_days)} gün")

# Gelecek tahminleri yap
future_predictions = model.predict(future_days)

print(f"✅ Gelecek tahminler tamamlandı!")

# Gelecek tarihleri oluştur
# Son tarihten 1 gün sonra başla, 30 gün devam et
last_date = daily_sales_ma['date'].max()
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30)

# Tahminleri DataFrame'e dönüştür
future_df = pd.DataFrame({
    'Tarih': future_dates,
    'Tahmin (TL)': future_predictions.round(2),
    'Gün': future_dates.day_name()  # Pazartesi, Salı, vb.
})

# Hafta bilgisi ekle (haftalık özet için)
future_df['Hafta_No'] = future_df['Tarih'].dt.isocalendar().week

print(f"\n📅 TAHMİN DÖNEMİ:")
print("="*70)
print(f"   • Başlangıç tarihi: {future_dates[0].strftime('%d %B %Y, %A')}")
print(f"   • Bitiş tarihi: {future_dates[-1].strftime('%d %B %Y, %A')}")
print(f"   • Toplam gün: 30")

# Özet istatistikler
gunluk_ort = future_predictions.mean()
haftalik_ort = gunluk_ort * 7
aylik_toplam = future_predictions.sum()
min_tahmin = future_predictions.min()
max_tahmin = future_predictions.max()

print(f"\n💰 TAHMİN ÖZETİ:")
print("="*70)
print(f"   • Günlük ortalama: {gunluk_ort:.2f} TL")
print(f"   • Haftalık beklenen: {haftalik_ort:.2f} TL")
print(f"   • Aylık toplam (30 gün): {aylik_toplam:.2f} TL")
print(f"   • Minimum tahmin: {min_tahmin:.2f} TL (En düşük gün)")
print(f"   • Maksimum tahmin: {max_tahmin:.2f} TL (En yüksek gün)")
print(f"   • Tahmin aralığı: {max_tahmin - min_tahmin:.2f} TL")

# Geçmiş ile karşılaştır
gecmis_ort = y_train.mean()
fark_yuzde = ((gunluk_ort - gecmis_ort) / gecmis_ort) * 100

print(f"\n📊 GEÇMİŞLE KARŞILAŞTIRMA:")
print("="*70)
print(f"   • Geçmiş ortalama (train): {gecmis_ort:.2f} TL")
print(f"   • Gelecek ortalama (tahmin): {gunluk_ort:.2f} TL")
print(f"   • Fark: {fark_yuzde:+.2f}%")

if fark_yuzde > 5:
    print(f"   → Satışlar artacak! 📈 Stok artırın!")
elif fark_yuzde < -5:
    print(f"   → Satışlar azalacak! 📉 Kampanya düşünün!")
else:
    print(f"   → Satışlar stabil kalacak ➡️")

# İlk 10 günün tahminlerini göster
print(f"\n📋 İLK 10 GÜNÜN TAHMİNLERİ:")
print("="*70)
print(future_df[['Tarih', 'Gün', 'Tahmin (TL)']].head(10).to_string(index=False))

# Haftalık özet oluştur
print(f"\n📅 HAFTALIK ÖZET:")
print("="*70)
haftalik_ozet = future_df.groupby('Hafta_No').agg({
    'Tahmin (TL)': ['sum', 'mean', 'count']
}).round(2)
haftalik_ozet.columns = ['Haftalık Toplam (TL)', 'Günlük Ortalama (TL)', 'Gün Sayısı']
print(haftalik_ozet.to_string())

print(f"\n✅ Gelecek tahminler hazır!")
```

```python
"""
========================================================================
BÖLÜM 13: EK GÖRSELLEŞTİRMELER
========================================================================
İş kararları için ek analizler:
1. Aylık karşılaştırma (Geçmiş vs Gelecek)
2. Haftalık trend analizi
3. Tahmin güven aralığı
========================================================================
"""

print("\n\n📊 EK GÖRSEL ANALİZLER OLUŞTURULUYOR...")
print("="*70)

# 1x3 grafik düzeni
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ========================================================================
# GRAFİK 1: Aylık Karşılaştırma (Sol)
# ========================================================================
# Geçmiş ayların ortalaması vs Gelecek tahmini
# İş için en önemli grafik: Hangi ay daha iyi?

print("\n📊 Aylık karşılaştırma hazırlanıyor...")

# Train verisini aylara göre grupla
train_data['month'] = train_data['date'].dt.to_period('M')
monthly_train = train_data.groupby('month')['revenue_ma7'].mean().reset_index()
monthly_train['month'] = monthly_train['month'].astype(str)
monthly_train['tip'] = 'Geçmiş (Gerçek)'

# Test verisini aylara göre grupla
test_data['month'] = test_data['date'].dt.to_period('M')
monthly_test = test_data.groupby('month')['revenue_ma7'].mean().reset_index()
monthly_test['month'] = monthly_test['month'].astype(str)
monthly_test['tip'] = 'Test (Gerçek)'

# Gelecek tahminleri aylara göre grupla
future_df['month'] = future_df['Tarih'].dt.to_period('M')
monthly_future = future_df.groupby('month')['Tahmin (TL)'].mean().reset_index()
monthly_future['month'] = monthly_future['month'].astype(str)
monthly_future.columns = ['month', 'revenue_ma7']
monthly_future['tip'] = 'Gelecek (Tahmin)'

# Hepsini birleştir
monthly_all = pd.concat([monthly_train, monthly_test, monthly_future])

# Bar chart çiz
x_positions = range(len(monthly_all))
colors_map = {'Geçmiş (Gerçek)': 'steelblue', 
              'Test (Gerçek)': 'green', 
              'Gelecek (Tahmin)': 'orange'}
colors = [colors_map[tip] for tip in monthly_all['tip']]

bars = axes[0].bar(x_positions, monthly_all['revenue_ma7'], 
                   color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)

# Her bar üzerine değer yaz
for i, (idx, row) in enumerate(monthly_all.iterrows()):
    axes[0].text(i, row['revenue_ma7'] + 50, f"{row['revenue_ma7']:.0f}", 
                ha='center', va='bottom', fontweight='bold', fontsize=9)

axes[0].set_xticks(x_positions)
axes[0].set_xticklabels(monthly_all['month'], rotation=45, ha='right')
axes[0].set_title('Aylık Ortalama Gelir Karşılaştırması', 
                 fontsize=12, fontweight='bold')
axes[0].set_ylabel('Ortalama Günlük Gelir (TL)', fontsize=11)
axes[0].set_xlabel('Ay', fontsize=11)
axes[0].grid(True, alpha=0.3, axis='y')

# Legend oluştur
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='steelblue', label='Geçmiş (Gerçek)'),
                   Patch(facecolor='green', label='Test (Gerçek)'),
                   Patch(facecolor='orange', label='Gelecek (Tahmin)')]
axes[0].legend(handles=legend_elements, loc='best', fontsize=9)

print("✅ Grafik 1 (Aylık Karşılaştırma): Oluşturuldu")

# ========================================================================
# GRAFİK 2: Haftalık Trend (Orta)
# ========================================================================
# Son 8 hafta + Gelecek 4 hafta trend'i
print("\n📊 Haftalık trend analizi hazırlanıyor...")

# Son 8 haftayı al
all_data = pd.concat([train_data, test_data])
all_data['week'] = all_data['date'].dt.to_period('W')
weekly_past = all_data.groupby('week')['revenue_ma7'].mean().tail(8).reset_index()
weekly_past['week'] = weekly_past['week'].astype(str)
weekly_past['tip'] = 'Geçmiş'

# Gelecek haftalık tahminler
weekly_future_data = future_df.groupby('Hafta_No')['Tahmin (TL)'].mean().reset_index()
weekly_future_data['week'] = ['H' + str(i) for i in weekly_future_data['Hafta_No']]
weekly_future_data = weekly_future_data[['week', 'Tahmin (TL)']]
weekly_future_data.columns = ['week', 'revenue_ma7']
weekly_future_data['tip'] = 'Gelecek'

# Birleştir
weekly_combined = pd.concat([weekly_past[['week', 'revenue_ma7', 'tip']], 
                             weekly_future_data])

# Çizgi grafik
past_mask = weekly_combined['tip'] == 'Geçmiş'
future_mask = weekly_combined['tip'] == 'Gelecek'

axes[1].plot(range(sum(past_mask)), 
            weekly_combined[past_mask]['revenue_ma7'].values,
            marker='o', linewidth=2.5, color='blue', 
            markersize=8, label='Geçmiş (8 hafta)')
axes[1].plot(range(sum(past_mask)-1, len(weekly_combined)), 
            [weekly_combined[past_mask]['revenue_ma7'].values[-1]] + 
            list(weekly_combined[future_mask]['revenue_ma7'].values),
            marker='D', linewidth=2.5, color='orange', 
            linestyle='--', markersize=8, label='Gelecek (Tahmin)')

axes[1].set_title('Haftalık Trend Analizi', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Haftalık Ortalama (TL)', fontsize=11)
axes[1].set_xlabel('Hafta', fontsize=11)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

# X ekseni etiketleri
all_weeks = list(weekly_combined['week'])
axes[1].set_xticks(range(len(all_weeks)))
axes[1].set_xticklabels([w[-5:] if len(w) > 5 else w for w in all_weeks], 
                       rotation=45, ha='right', fontsize=9)

print("✅ Grafik 2 (Haftalık Trend): Oluşturuldu")

# ========================================================================
# GRAFİK 3: Tahmin Güven Aralığı (Sağ)
# ========================================================================
# Gelecek tahminler için üst ve alt sınırlar
# İstatistiksel güven aralığı: Tahmin ± (MAE × 1.5)
print("\n📊 Güven aralığı hesaplanıyor...")

# Güven aralığını hesapla
# MAE'yi kullanarak ±%95 güven aralığı
confidence_margin = mae * 1.96  # 1.96 = %95 güven için z-skoru

future_lower = future_predictions - confidence_margin
future_upper = future_predictions + confidence_margin

# Ana tahmin çizgisi
axes[2].plot(range(30), future_predictions, 
            color='darkblue', linewidth=3, 
            marker='o', markersize=6, label='Tahmin')

# Güven aralığı (gölgeli alan)
axes[2].fill_between(range(30), future_lower, future_upper,
                     alpha=0.3, color='lightblue', 
                     label=f'Güven Aralığı (%95)')

# Üst ve alt sınır çizgileri
axes[2].plot(range(30), future_upper, 
            color='red', linewidth=1.5, 
            linestyle='--', alpha=0.7, label='Üst Sınır')
axes[2].plot(range(30), future_lower, 
            color='green', linewidth=1.5, 
            linestyle='--', alpha=0.7, label='Alt Sınır')

axes[2].set_title('Gelecek Tahminler - Güven Aralığı', 
                 fontsize=12, fontweight='bold')
axes[2].set_xlabel('Gün (1-30)', fontsize=11)
axes[2].set_ylabel('Gelir (TL)', fontsize=11)
axes[2].legend(fontsize=9, loc='best')
axes[2].grid(True, alpha=0.3)

# İstatistikler ekle
stats_text = f'Tahmin: {future_predictions.mean():.0f} TL\n'
stats_text += f'Alt: {future_lower.mean():.0f} TL\n'
stats_text += f'Üst: {future_upper.mean():.0f} TL'
axes[2].text(0.02, 0.98, stats_text,
            transform=axes[2].transAxes, fontsize=9,
            verticalalignment='top', 
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))

print("✅ Grafik 3 (Güven Aralığı): Oluşturuldu")

plt.tight_layout()
plt.show()

print("\n✅ Tüm ek grafikler oluşturuldu!")

print("\n💡 GRAFİK YORUMLARI:")
print("="*70)
print("📊 Grafik 1 (Aylık Karşılaştırma):")
print("   • Hangi ay daha iyi performans gösterdi?")
print("   • Gelecek ay geçmişe göre nasıl?")
print("   • Mevsimsel patern var mı?")

print("\n📊 Grafik 2 (Haftalık Trend):")
print("   • Son haftalarda trend yukarı mı aşağı mı?")
print("   • Gelecek haftalarda nasıl devam edecek?")
print("   • Ani değişimler var mı?")

print("\n📊 Grafik 3 (Güven Aralığı):")
print("   • Mavi alan: %95 olasılıkla bu aralıkta olacak")
print("   • Geniş alan: Belirsizlik yüksek")
print("   • Dar alan: Tahmin güvenilir")
print(f"   • Sizin durumunuzda: ±{confidence_margin:.0f} TL belirsizlik")
```
