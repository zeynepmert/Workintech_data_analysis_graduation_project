```python
import numpy as np
import pandas as pd
import os
print(os.getcwd())
```

```python
df = pd.read_csv("../data/sales.csv")
df
```

```python
print(df.isnull().sum())
#eksik veri olup olmadığını kontrol ediyoruz
```

```python
print(df.dtypes)
# sütun formatlarını kontrol ediyoruz
```

```python
df = df[df["price_per_unit"] > 0]
# Negatif veya sıfır fiyat var mı kontrol ediyoruz
```

```python
df = df[df["quantity"] > 0]
#sıfır adet ürün görünüyor mu kontrol ediyoruz
```

```python
print(df.describe())
print(df.info())
#verileri inceliyoruz
```

```python
ortalama_fiyat = df["price_per_unit"].mean()
print(f"Ortalama Birim Fiyat: ₺{ortalama_fiyat:.2f}")
```

```python
ortalama_adet = df["quantity"].mean()
print(f"Ortalama Sepet Adedi: {ortalama_adet:.2f} ürün")
```

```python
ortalama_tutar = df["total_price"].mean()
print(f"Ortalama Sepet Tutarı: ₺{ortalama_tutar:.2f}")
```

```python
siparis_bazli = df.groupby("order_id")["total_price"].sum()
print(f"Ortalama Sipariş Tutarı: ₺{siparis_bazli.mean():.2f}")
```
