```python
import pandas as pd
```

```python
churn_df = pd.read_csv('../data/orders_with_churn.csv')
payment_df = pd.read_csv('../data/orders_with_payment.csv')

churn_df.head()
```

```python
payment_df.head()
```

```python
# Drop the 'payment' column from churn_df
churn_df = churn_df.drop('payment', axis=1)

# Merge the 'churn' column from churn_df to payment_df based on 'order_id'
merged_df = pd.merge(payment_df, churn_df[['order_id', 'churn']], on='order_id', how='left')

# Display the merged DataFrame
display(merged_df.head())
```

```python
merged_df.to_csv('orders_with_churn_and_payment.csv', index=False)
```
