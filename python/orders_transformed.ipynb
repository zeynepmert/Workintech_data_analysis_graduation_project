### orders_update Duzeltilmis Data
Orders_update.csv dosyasindaki payment degerleri yanlis oldugu icin bunlari total_price lari hesaplayarak yeni bir payment sutunu ekledim. Eski (hatali) payment sutununu sildim ve bunu 'orders_with_payment.csv' olarak export ettim.

```python
import pandas as pd
```

```python
orders_df = pd.read_csv('../data/orders_updated.csv')
sales_df = pd.read_csv('../data/sales.csv')
```

```python
# payment sutununu silme
orders_df = orders_df.drop('payment', axis=1)
```

```python
#
order_total_price = sales_df.groupby('order_id')['total_price'].sum().reset_index()
```

```python
# total_price ekleme
orders_df = pd.merge(orders_df, order_total_price, on='order_id', how='left')
orders_df = orders_df.rename(columns={'total_price': 'payment'})
orders_df.head()
```

```python
# csv olarak disa aktarma
orders_df.to_csv('orders_with_payment.csv', index=False)
```
